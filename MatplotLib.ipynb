### MATPLOYLIB 
Matplotlib is a plotting library for the python programming language.

Some of the major pros of Matplotlib are:
1. Generally easy to get started for simple plot
2. Support for custom labels and texts
3. Great control for every element in a figure
4. High-Quality outputs in many formats
5. very customizable in general





```python
import matplotlib.pyplot as plt

# below statement is used to avoid plt.show() command for displaying the graph. 
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.arange(0,10)
y=np.arange(11,21)
```

# Scatter Plot

```python
plt.scatter(x,y, c='r' )
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Graph')
```

# Plt plot

```python
y=x*x
```

```python
plt.plot(x,y,'g*--', linestyle='dashed', linewidth=2, markersize=12)
```

# Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
np.pi
```

```python
x=np.arange(0,5*np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'g--')
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos, 'b--')
plt.title('cosine')
```

# Bar chart

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[10,15,14]

plt.bar(x,y, align='center')
plt.bar(x2,y2,color='g')
plt.title('bar graph')
plt.ylabel('Y axis')
plt.xlabel('x Axis')
```

# Histograms

```python
q = np.array([10,14,18,19,6,29,32,8,44,69,1,21,33])
plt.hist(q)
plt.title('histogram')
```

# Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
```

```python
plt.boxplot(data, vert=True, patch_artist=True)
```
